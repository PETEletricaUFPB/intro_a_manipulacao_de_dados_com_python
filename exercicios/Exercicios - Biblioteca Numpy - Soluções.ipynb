#### Importe NumPy como np

```python
import numpy as np
```

#### Crie uma matriz de 10 zeros

```python
np.zeros(10)
```

#### Crie uma matriz de 10 ones

```python
np.ones(10)
```

#### Crie uma matriz de 10 cincos

```python
np.ones(10) * 5
```

#### Crie um array de inteiros de 10 at 50

```python
np.arange(10,51)
```

#### Crie um array dos numeros pares de 10 at 50

```python
np.arange(10,51,2)
```

#### Criei uma matriz 3x3 com valores variando de 0 at 8

```python
np.arange(9).reshape(3,3)
```

#### Crie uma matriz identidade 3x3

```python
np.eye(3)
```

#### Use NumPy para gerar nmeros aleatrios entre 0 e 1

```python
np.random.rand(1)
```

#### Use Numpy para gerar um array de 25 nmeros aleatrios tirados de uma distribuio normal.

```python
np.random.randn(25)
```

#### Crie a seguinte matriz:

```python
np.arange(1,101).reshape(10,10) / 100
```

#### Crie um array de tamanho 20 igualmente espaado entre 0 e 1.

```python
np.linspace(0,1,20)
```

## Indexao Numpy e Seleo

Agora voc receber algumas matrizes e ser solicitado a replicar as sadas resultantes da matriz:

```python
mat = np.arange(1,26).reshape(5,5)
mat
```

```python
# ESCREVE O CDIGO AQUI QUE REPRODICA A SADA DA CLULA ABAIXO
# SEJA CUIDADOSO PARA NO RODAR A CLULA ABAIXO, DE OUTRA FORMA, NO PODER
# PODE VER A MATRIZ DESEJADA
```

```python
mat[2:,1:]
```

```python
# ESCREVE O CDIGO AQUI QUE REPRODICA A SADA DA CLULA ABAIXO
# SEJA CUIDADOSO PARA NO FUNCIONAR A CLULA ABAIXO, DE OUTRA FORMA, NO PODER
# PODE VER A SORTEZA MAIS QUALQUER
```

```python
mat[3,4]
```

```python
# ESCREVE O CDIGO AQUI QUE REPRODICA A SADA DA CLULA ABAIXO
# SEJA CUIDADOSO PARA NO FUNCIONAR A CLULA ABAIXO, DE OUTRA FORMA, NO PODER
# PODE VER A SORTEZA MAIS QUALQUER
```

```python
mat[:3,1:2]
```

```python
# ESCREVE O CDIGO AQUI QUE REPRODICA A SADA DA CLULA ABAIXO
# SEJA CUIDADOSO PARA NO FUNCIONAR A CLULA ABAIXO, DE OUTRA FORMA, NO PODER
# PODE VER A SORTEZA MAIS QUALQUER
```

```python
mat[4,:]
```

```python
# ESCREVE O CDIGO AQUI QUE REPRODICA A SADA DA CLULA ABAIXO
# SEJA CUIDADOSO PARA NO FUNCIONAR A CLULA ABAIXO, DE OUTRA FORMA, NO PODER
# PODE VER A SORTEZA MAIS QUALQUER
```

```python
mat[3:5,:]
```
