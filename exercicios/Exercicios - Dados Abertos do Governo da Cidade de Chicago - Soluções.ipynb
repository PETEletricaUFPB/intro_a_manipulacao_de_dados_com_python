## Importe pandas e leia no arquivo csv "salarios" e configure-o para um DataFrame chamado df.

```python
import pandas as pd
```

```python
df = pd.read_csv('salarios_exercicio.csv')
```

## Verifique o "head" do DataFrame.

```python
df.head()
```

## Quantas linhas e colunas existem?

```python
df.info()
```

## Qual é o preço médio dos salários?

```python
df['Employee Annual Salary'].mean()
```

## Qual a mediana dos salários?

```python
df['Employee Annual Salary'].median()
```

## Qual o valor do salário mais altos e mais baixos?

```python
df['Employee Annual Salary'].max()
```

```python
df['Employee Annual Salary'].min()
```

## Quais são os 10 maiores salários?

```python
maiores_salarios = df['Employee Annual Salary'].sort_values(ascending=False)
maiores_salarios[:10]
```

## Quais são os 5 títulos de trabalho mais comuns?

```python
trabalhos_mais_comuns = df['Department'].value_counts()
trabalhos_mais_comuns[:5]
```

## Quantas pessoas são policiais?

```python
df[df['Department'] == "POLICE"].count()
```

```python
df['Department'].value_counts()
```

## Crie pelo menos 1 gráfico com as informações do DataFrame

```python
maiores_salarios[:10].plot(
    kind='bar',
    figsize=(16,8),
    title='Dez maiores salários por indice do DataFrame',
    ylabel='Salário',
    xlabel='Índice no DataFrame',
    grid='on'
)
```

```python
maiores_salarios.describe().plot(kind='pie',figsize=(16,8))
```

```python
trabalhos_mais_comuns[:5].plot(
    kind='line',
    figsize=(16,8),
    title='Cinco primeiros departamentos com mais funcionários',
    ylabel='Quantidade de Funcionários',
    xlabel='Departamento',
    grid='on'
    )
```

```python
df['Department'].value_counts().plot(figsize=(16,8))
```
