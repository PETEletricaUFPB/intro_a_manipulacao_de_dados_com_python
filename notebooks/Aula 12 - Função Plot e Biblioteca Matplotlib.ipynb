<h2 align="left" style="color:blue">
    Informaes sobre a funo plot()
</h2>

<hr align="left" width="400px">

 * A biblioteca Pandas facilita o trabalho de criar os grficos encapsulando a biblioteca Matplotlib;
    * Quando voc no estiver com a biblioteca Matplotlib instalado no seu ambiente virtual, o seguinte erro ser retornado:
    > `ImportError: matplotlib is required for plotting when the default backend "matplotlib" is selected.`
 * Para criar os grficos precisamos apenas chamar a funo `plot()`;
 * Por padro a biblioteca Pandas ir exibir os dados como um grfico de linha.

    [Chart Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/boletim_diario.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# organizando o df a partir dos valores da coluna data
df.sort_values(by='data',inplace=True)
df.head()
```

```python
df1 = df.filter(items=['casosNovos','obitosNovos','data']) # filtrando os dados com base em 3 colunas
df1.set_index('data',inplace=True) # o novo dataframe df1 teve seu index ajustado definitivamente

# visualizao rpida com a funo plot
df1[:50].plot.bar(
    title='Grfico com quantidade de Novos Casos e Novos Obitos durante os 50 primeiros dias da pandemia - PB',
    ylabel='casosNovos e obitosNovos',
    xlabel='data',
    figsize=(16,8)
)
```

**Obs.:** Quando necessitamos de uma visualizao rpida sem a necessidade de modificar a parte esttica do grfico,  possvel utilizar a funo `plot()` da biblioteca pandas. Na situao acima, um grafico ao lado do outro (subplots) pode ser gerado e melhor confeccionado para apresentao das informaes como no exemplo a seguir.

<h2 align="left" style="color:blue">
    Subplots com a Biblioteca Matplotlib
</h2>

<hr align="left" width="400px">

```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(18,8))

ax1 = df1['casosNovos'].plot(
    color='red',
    ax=ax1,
    title='Grfico com quantidade de Novos Casos e Novos Obitos - 313 primeiros dias',
)

ax2 = df1['obitosNovos'].plot(
    color='blue',
    ax=ax2,
    title='Grfico com quantidade de Novos Casos e Novos Obitos - 313 primeiros dias',
)
```

<h2 align="left" style="color:blue">
    Subplots com a Biblioteca Pandas
</h2>

<hr align="left" width="400px">

```python
df1.plot(
    title='Grfico com quantidade de Novos Casos e Novos Obitos durante os 50 primeiros dias da pandemia - PB',
    ylabel='Quantidade',
    xlabel='Data',
    subplots=True,
    grid=True,
    figsize=(16,8)
)
```

<h2 align="left" style="color:blue">
    Grfico de rea com a funo Plot()
</h2>

<hr align="left" width="400px">

```python
df2 = df.filter(items=['descartados','data'])
df2.set_index('data',inplace=True)

df2.plot(kind='area',figsize=(16,8))
```

<h2 align="left" style="color:blue">
    Outros tipos de plots
</h2>

<hr align="left" width="400px">

Utilizando o parmetro `kind`,  possvel alterar entre os tipos de grficos abaixo:
* `bar` ou  `barh` para um grfico de barras
* `hist` para um histograma
* `box` para um boxplot
* `kde` ou `density` para um grfico de densidade
* `area` para um grfico de rea
* `scatter` para um grfico de pontos
* `hexbin` para um grfico hexagonais
* `pie` para um grfico de pizza

```python
df4 = df[:50].filter(items=['descartados'])
# df4.set_index('data',inplace=True)
```

<h3 align="left" style="color:blue">
    Grfico de barras
</h3>

```python
df4.plot(kind="bar",figsize=(16,8))
```

<h3 align="left" style="color:blue">
    Grfico de Barras Empilhado
</h3>

```python
# df4.plot.bar()  igual a df4.plot(kind='bar')
df4.plot.bar(figsize=(16,8),stacked=True)
```

<h3 align="left" style="color:blue">
    Grfico Histograma
</h3>

```python
df4.plot.hist(
    stacked=True,
    bins=20,
    figsize=(16,8)
);
```

<h3 align="left" style="color:blue">
    Grfico de Pontos
</h3>

```python
df5 = df.filter(items=['descartados'])
df5.reset_index(inplace=True)
```

```python
df5.plot.scatter(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```

```python
df5[:100].plot.scatter(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```

```python
df5[:100].plot.line(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```
