<h2 align="left" style="color:blue">
    Informações sobre a função plot()
</h2>

<hr align="left" width="400px">

 * A biblioteca Pandas facilita o trabalho de criar os gráficos encapsulando a biblioteca Matplotlib;
    * Quando você não estiver com a biblioteca Matplotlib instalado no seu ambiente virtual, o seguinte erro será retornado:
    > `ImportError: matplotlib is required for plotting when the default backend "matplotlib" is selected.`
 * Para criar os gráficos precisamos apenas chamar a função `plot()`;
 * Por padrão a biblioteca Pandas irá exibir os dados como um gráfico de linha.

    [Chart Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/boletim_diario.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# organizando o df a partir dos valores da coluna data
df.sort_values(by='data',inplace=True)
df.head()
```

```python
df1 = df.filter(items=['casosNovos','obitosNovos','data']) # filtrando os dados com base em 3 colunas
df1.set_index('data',inplace=True) # o novo dataframe df1 teve seu index ajustado definitivamente

# visualização rápida com a função plot
df1[:50].plot.bar(
    title='Gráfico com quantidade de Novos Casos e Novos Obitos durante os 50 primeiros dias da pandemia - PB',
    ylabel='casosNovos e obitosNovos',
    xlabel='data',
    figsize=(16,8)
)
```

**Obs.:** Quando necessitamos de uma visualização rápida sem a necessidade de modificar a parte estética do gráfico, é possível utilizar a função `plot()` da biblioteca pandas. Na situação acima, um grafico ao lado do outro (subplots) pode ser gerado e melhor confeccionado para apresentação das informações como no exemplo a seguir.

<h2 align="left" style="color:blue">
    Subplots com a Biblioteca Matplotlib
</h2>

<hr align="left" width="400px">

```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(18,8))

ax1 = df1['casosNovos'].plot(
    color='red',
    ax=ax1,
    title='Gráfico com quantidade de Novos Casos e Novos Obitos - 313 primeiros dias',
)

ax2 = df1['obitosNovos'].plot(
    color='blue',
    ax=ax2,
    title='Gráfico com quantidade de Novos Casos e Novos Obitos - 313 primeiros dias',
)
```

<h2 align="left" style="color:blue">
    Subplots com a Biblioteca Pandas
</h2>

<hr align="left" width="400px">

```python
df1.plot(
    title='Gráfico com quantidade de Novos Casos e Novos Obitos durante os 50 primeiros dias da pandemia - PB',
    ylabel='Quantidade',
    xlabel='Data',
    subplots=True,
    grid=True,
    figsize=(16,8)
)
```

<h2 align="left" style="color:blue">
    Gráfico de Área com a função Plot()
</h2>

<hr align="left" width="400px">

```python
df2 = df.filter(items=['descartados','data'])
df2.set_index('data',inplace=True)

df2.plot(kind='area',figsize=(16,8))
```

<h2 align="left" style="color:blue">
    Outros tipos de plots
</h2>

<hr align="left" width="400px">

Utilizando o parâmetro `kind`, é possível alterar entre os tipos de gráficos abaixo:
* `bar` ou  `barh` para um gráfico de barras
* `hist` para um histograma
* `box` para um boxplot
* `kde` ou `density` para um gráfico de densidade
* `area` para um gráfico de área
* `scatter` para um gráfico de pontos
* `hexbin` para um gráfico hexagonais
* `pie` para um gráfico de pizza

```python
df4 = df[:50].filter(items=['descartados'])
# df4.set_index('data',inplace=True)
```

<h3 align="left" style="color:blue">
    Gráfico de barras
</h3>

```python
df4.plot(kind="bar",figsize=(16,8))
```

<h3 align="left" style="color:blue">
    Gráfico de Barras Empilhado
</h3>

```python
# df4.plot.bar() é igual a df4.plot(kind='bar')
df4.plot.bar(figsize=(16,8),stacked=True)
```

<h3 align="left" style="color:blue">
    Gráfico Histograma
</h3>

```python
df4.plot.hist(
    stacked=True,
    bins=20,
    figsize=(16,8)
);
```

<h3 align="left" style="color:blue">
    Histograma de Duas Variáveis
</h3>

```python
df[['casosAcumulados','descartados']].plot.hist(
    stacked=True,
    #bins=20,
    figsize=(16,8)
);
```

```python
df[['casosAcumulados','descartados']].plot.hist(
    stacked=True,
    orientation="horizontal",
    figsize=(16,8)
)
```

<h3 align="left" style="color:blue">
    Gráfico de Pontos
</h3>

```python
df5 = df.filter(items=['descartados'])
df5.reset_index(inplace=True)
```

```python
df5.plot.scatter(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```

```python
df5[:100].plot.scatter(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```

```python
df5[:100].plot.line(
    x='index',
    y='descartados',
    figsize=(16,8)
)
```
