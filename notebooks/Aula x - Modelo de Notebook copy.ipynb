<h2 align="left" style="color:blue">
    Informações sobre a função plot()
</h2>

<hr align="left" width="400px">

 * A biblioteca Pandas facilita o trabalho de criar os gráficos encapsulando a biblioteca Matplotlib.
    * Quando você não estiver com a biblioteca Matplotlib instalado no seu ambiente virtual, o seguinte erro será retornado:
    > `ImportError: matplotlib is required for plotting when the default backend "matplotlib" is selected.`
 * Para criar os gráficos precisamos apenas chamar a função `plot()`, por padrão a biblioteca Pandas irá exibir os dados como um gráfico de linha.

 [Chart Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/boletim_diario.csv')
df.head()
```

```python
df.info()
```

```python
df[:50].filter(items=['casosNovos','obitosNovos','data']).set_index('data').plot.bar(figsize=(18,8))
```

<h2 align="left" style="color:blue">
    Outros tipos de plots
</h2>

<hr align="left" width="400px">

Utilizando o parâmetro `kind`, é possível alterar entre os tipos de gráficos abaixo:
* `bar` ou  `barh` para um gráfico de barras
* `hist` para um histograma
* `box` para um boxplot
* `kde` ou `density` para um gráfico de densidade
* `area` para um gráfico de área
* `scatter` para um gráfico de pontos
* `hexbin` para um gráfico hexagonais
* `pie` para um gráfico de pizza
